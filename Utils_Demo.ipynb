```python
from scipy.io import loadmat 

train_svhn = loadmat('data/shvn/train.mat')
test_shvn = loadmat('data/shvn/test.mat')

X_train, y_train = train_svhn['X'], train_svhn['y']
X_test, y_test = test_shvn['X'], test_shvn['y']

print("Shape of X_train ", X_train.shape)
print("Shape of X test ", X_test.shape)

# roll axis backward

X_train = np.rollaxis(X_train, 3)
X_test  = np.rollaxis(X_test, 3)
print("Reshaped X_train to ", X_train.shape)
print("Reshaped X test to ", X_test.shape)


```
